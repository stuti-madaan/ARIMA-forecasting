```python
import pandas as pd 
import pmdarima as pm
import math
import numpy as np 
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from sklearn.metrics import mean_squared_error 
```

```python
data=pd.read_csv("./H1.csv")
```

```python
d1=data[['ArrivalDateYear','ArrivalDateWeekNumber','ADR']]
d1['ArrivalDateYearWeek'] = d1.apply(lambda r: str(int(r.ArrivalDateYear))+str(int(r.ArrivalDateWeekNumber)).zfill(2), axis=1)
d1.sort_values(['ArrivalDateYear','ArrivalDateWeekNumber'], inplace=True)#.reset_index(inplace=True)
d1_grouped = d1.groupby(['ArrivalDateYearWeek']).agg({'ADR':'mean'})
plt.plot(d1_grouped.index, d1_grouped['ADR'])
```

```python
d2=pd.read_csv("./H2.csv")
d2=d2[['ArrivalDateYear','ArrivalDateWeekNumber','ADR']]
d2['ArrivalDateYearWeek'] = d2.apply(lambda r: str(int(r.ArrivalDateYear))+str(int(r.ArrivalDateWeekNumber)).zfill(2), axis=1)
d2.sort_values(['ArrivalDateYear','ArrivalDateWeekNumber'], inplace=True)#.reset_index(inplace=True)
d2_grouped = d2.groupby(['ArrivalDateYearWeek']).agg({'ADR':'mean'})
plt.plot(d2_grouped.index, d2_grouped['ADR'])
```

```python
train_df = d1_grouped.iloc[:100,:]
plot_acf(train_df, lags=60, zero=False);

## Seasonality = 52 Weeks 
```

## ARIMA Model Training ## 

```python
m=52
```

- Auto-Regressive (p) -> Number of autoregressive terms.
- Integrated (d) -> Number of nonseasonal differences needed for stationarity.
- Moving Average (q) -> Number of lagged forecast errors in the prediction equation.

In the Auto ARIMA model, small p,d,q values represent non-seasonal components, and capital P, D, Q represent seasonal components. It works similarly like hyper tuning techniques to find the optimal value of p, d, and q with different combinations and the final values would be determined with the lower AIC, BIC parameters taking into consideration.

```python
## Test For Stationarity ## 
adf_test=pm.arima.ADFTest(alpha=0.05)
adf_test.should_diff(train_df['ADR'])
```

#### Non Stationary Data

```python
Arima_model=pm.auto_arima(train_df, start_p=0, start_q=0, max_p=10, max_q=10, \
                          start_P=0, start_Q=0, max_P=10, max_Q=10, m=52, stepwise=True, \
                          seasonal=True, information_criterion='aic', trace=True, d=1, D=1, \
                          error_action='warn', suppress_warnings=True, random_state = 20, n_fits=30)
```

```python
test_df=d1_grouped.iloc[100:,:]['ADR']
```

```python
## Predictions on Test DataSet ## 
predictions=pd.DataFrame(Arima_model.predict(n_periods=15), index=test_df)
predictions=np.array(predictions)
```

```python
plt.plot(predictions)
plt.plot(test_df)
```

```python
## RMSE ##
```

```python
mse = mean_squared_error(test_df, predictions)
rmse = math.sqrt(mse)
print('RMSE: %f' % rmse)
np.mean(test_df)
```

```python
plt.figure(figsize=(16,8))
plt.plot(train_df.index, train_df['ADR'].values, label='train')
plt.plot(test_df.index, test_df.values, label='test')
plt.plot(test_df.index, [i[0] for i in predictions], label='prediction')
plt.xlabel("Year+Week")
plt.xticks(rotation='vertical')
plt.ylabel("Avg Weekly ADR")
plt.title("ADR v/s Week - Actual and Predictions")
plt.legend(loc='best')
plt.show()
```
